```python
def rename_columns_by_range(df, start_column_name, end_column_name, new_start_year):
    # Get the starting and ending column indices
    start_column_index = df.columns.get_loc(start_column_name)
    end_column_index = df.columns.get_loc(end_column_name)

    # Create the new column names as a range
    new_column_names = list(range(new_start_year, new_start_year + (end_column_index - start_column_index + 1)))

    # Rename columns
    df.columns = [*df.columns[:start_column_index], *new_column_names, *df.columns[end_column_index + 1:]]

# Example usage
rename_columns_by_range(df, 'Unnamed: 4', 'Unnamed: 66', 1960)

# Display the DataFrame
print(df)
```

```python
df['country'] = df['country'].replace({'Israel': 'palestine'})
df['country'] = df['country'].replace({"Cote d'Ivoire": 'Cote dIvoire'})
df['country'] = df['country'].replace({'Congo, Dem. Rep.': 'Congo Dem Rep'})
df['country'] = df['country'].replace({'Congo, Rep.': 'Congo Rep'})
```

```python
sub_saharan_african_countries = [
    'Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 'Cabo Verde',
    'Cameroon', 'Central African Republic', 'Chad', 'Comoros',"Cote d'Ivoire"
    'Democratic Republic of the Congo', 'Djibouti', 'Equatorial Guinea',
    'Eritrea', 'Eswatini', 'Ethiopia', 'Gabon', 'The Gambia', 'Ghana',
    'Guinea', 'Guinea-Bissau', 'Ivory Coast', 'Kenya', 'Lesotho',
    'Liberia', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Mauritius',
    'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Republic of the Congo',
    'Rwanda', 'São Tomé and Príncipe', 'Senegal', 'Seychelles',
    'Sierra Leone', 'Somalia', 'South Africa', 'South Sudan', 'Sudan',
    'Tanzania', 'Togo', 'Uganda', 'Zambia', 'Zimbabwe','Congo Rep', 'Congo Dem Rep','Cote dIvoire'
]
```

```python
# Add a new column 'Region' based on Sub-Saharan Africa countries
df['Region'] = df['country'].apply(lambda x: 'Sub-Saharan' if x in sub_saharan_african_countries else 'Other')
```

```python
# Merge data with GeoJSON file
merged_data = africa_geojson.merge(df, left_on='name', right_on='Region', how='left')
```

```python
# Filter data for Sub-Saharan Africa and the year 2015
filtered_data = df[(df['Region'] == 'Sub-Saharan') & (df['Year'] == 2015)]
```

```python
filtered_data
```

```python
# Create an interactive choropleth map for Africa
fig = px.choropleth(
    filtered_data,
    locations='country', #Specifies the column in filtered_data that contains the country names.
    locationmode='country names', #Indicates that the location names are provided as actual country names.
    color='Value', #Specifies the column in filtered_data that contains the values to be used for coloring the map
    color_continuous_scale="Viridis", #Specifies the color scale to be used for the choropleth map.
    title= 'Adolescent Fertility Rate in Sub-Saharan Africa (2015)',
    labels={'Value': 'Births per 1,000 women aged 15-19'},
    hover_name='country',
    animation_frame='Year',
    projection='natural earth',
    template='plotly',
)

# Set color scale range and midpoint
color_scale_range = [50, 200]



# Show the interactive map
fig.show()
```
